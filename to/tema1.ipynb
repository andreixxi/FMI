```python
import numpy as np
import matplotlib.pyplot as plt
```

# problema maximizare
## 1) sistem cu solutie optima finita
sup(x1 + 2x2)<br>
conditiile:<br> 
-x1 + x2 <= 4 <br>
x1 + x2 <= 8<br>
<!-- x1, x2 >= 0<br> -->

```python
x1 = np.linspace(-5, 10, 200) # discretizare

plt.xlim(-5, 10)
plt.ylim(-1, 9)

# -x1 + x2 <= 4
x2 = 4 + x1
plt.plot(x1, x2)
plt.fill_between(x1, x2, x2 -15, alpha=0.2)

# x1 + x2 <= 8
x2 = 8 - x1
plt.plot(x1, x2)
plt.fill_between(x1, x2, x2 -15, alpha=0.2)

plt.scatter(2, 6, c='red', zorder=5) # punctul de optim
# cea mai mare valoare a lui x1 + 2x2 se obtine in punctul de intersectie ale dreptelor -x1 + x2 = 4 si x1 + x2 = 8

plt.show()
```

## 2) sistem cu optim unic, infinit
sup(x1 + 2x2)<br>
conditiile:<br> 
-x1 + x2 <= 4 <br>
x1 + 2x2 <= 8<br>
<!-- x1, x2 >= 0<br> -->

```python
x1 = np.linspace(-5, 10, 200) # discretizare

plt.xlim(-5, 10)
plt.ylim(-1, 9)

# -x1 + x2 <= 4
x2 = 4 + x1
plt.plot(x1, x2)
plt.fill_between(x1, x2, x2 -15, alpha=0.2)

# x1 + 2x2 <= 8
x2 = (8 - x1) / 2
plt.plot(x1, x2)
for i in range(0, len(x1), 10):
    plt.scatter(x1[i], x2[i])
plt.fill_between(x1, x2, x2 -15, alpha=0.2)

plt.show()
## toate punctele de pe dreapta portocalie sunt sol optime
```

## 3) sistem cu optim infinit
sup(x1 + 2x2)<br>
conditiile:<br> 
-x1 + x2 <= 4 <br>
x1 + 2x2 >= 8<br>
<!-- x1, x2 >= 0<br> -->

```python
x1 = np.linspace(-5, 10, 100) # discretizare

plt.xlim(-5, 10)
plt.ylim(-5, 10)

# x1 + 2x2 >= 3
x2 = 4+x1
plt.plot(x1, x2)

# x1 - 2x2 >= 2
x3 = (8 - x1)/2
plt.plot(x1, x3)

plt.fill_between(x1, x2, x3, where = x2>x3, alpha = 0.2)

plt.show()
## punctele din zona hasurata sunt sol optime
```

## 4) sistem incompatibil
sup(x1 + x2)<br>
conditiile:<br> 
x1 + 2x2 >= 3<br>
x1 - 2x2 >= 2<br>
x2 <= -2

```python
x1 = np.linspace(-5, 5, 100) # discretizare

plt.xlim(-3, 5)
plt.ylim(-4, 6)

#x1 + 2x2 >= 3
x2 = (3 - x1)/2
plt.plot(x1, x2)
plt.fill_between(x1, x2, x2 + 10, alpha=0.2)

# x1 - 2x2 >= 2
x3 = (x1 - 2)/2
plt.plot(x1, x3)
plt.fill_between(x1, x3, x3 + 10, alpha=0.2)

# x2 <= -2
x4 = x1 * 0 - 2
plt.plot(x1, x4)
plt.fill_between(x1, x4, x4 - 10, alpha=0.2)

plt.show()

## nu exista o zona comuna celor 3 inecuatii
```
