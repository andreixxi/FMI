```python
np.zeros((4, 4)) # cu valori de 0
np.ones((4, 4)) # cu valori de 1
np.full((4, 4), 9) # cu valori de 9
np.eye(4) # matricea I4
np.random.random((4, 4)) * 100 # 4 * 4 valori random in int [0, 100)
n = np.random.normal(100, 15, (4, 4)) # normal (Gaussian) distribution
                                  # mean = 100, std deviation = 15
r = np.random.random((4, 4)) * 10 # 4 * 4 valori random in int [0, 10)
# print(r)
r[1:3, 1:3] # elementele de pe liniile [1, 3), coloanele [1, 3)
r[3, 2] # linia 3 coloana 2
np.ravel(r) # A 1-D array, containing the elements of the input, is returned

r.reshape((2, 2, 2, 2))
bool_idx = (r > 3) # elementele din r mai mari decat 3 -> True 
r[bool_idx] # afiseaza r[True]
```

```python
r + n # np.add(r, n)
r - n # np.subtract(r, n)
r * n # np.multiply(r, n)
r / n # np.divide(r, n)
np.dot(r, n) # XxY YxZ
np.matmul(r, n) #Matrix product of two arrays
```

```python
print(r)
print(r.T) # The transposed array
print(np.linalg.inv(r)) # Compute the inverse of a matrix
```

```python
np.sum(r) # Axis or axes along which a sum is performed. The default, axis=None, will sum all of the elements 
          # of the input array. If axis is negative it counts from the last to the first axis.
    
np.sum(r, axis = 0) # suma pe coloane
np.sum(r, axis = 1) # suma pe linii

np.sum(r, axis = (0, 1)) # If axis is a tuple of ints, a sum is performed on all of the axes 
                         # specified in the tuple instead of a single axis or all the axes as before.
                         # in cazul asta, suma pe linii si coloane (toata matricea)
        
print(np.mean(r)) # media
print(np.std(r))  # deviatia standard

# pe linii
print("linii")
print(np.mean(r, axis = 1))
print(np.std(r, axis = 1))

# pe coloane
print("coloane")
print(np.mean(r, axis=0))
print(np.std(r, axis=0))
```

```python
print(r)
print(np.argmax(r)) # pozitia
print(np.max(r))    # valoarea

# de pe linii
print(np.argmax(r, axis = 1)) # maximul de pe fiecare linie
print(np.max(r, axis = 1))    # indexul maximului de pe fiecare linie

# de pe coloane
print(np.argmax(r, axis = 0))
print(np.max(r, axis = 0))
```

```python
import matplotlib.pyplot as plt
plt.rcParams.size = [9, 9] # dimensiunea figurii din plot

x = np.arange(0, 3 * np.pi, 0.1)  # step = 0.1
y = np.sin(x)

print(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Titlu')

plt.legend(['Sin'])
plt.show()
```

```python
z = np.cos(x)

plt.plot(x, y, color = 'red') # y = sin(x)   
plt.plot(x, z, color = 'green') # z = cos(x)

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('Titlu')

plt.legend(['Sin', 'Cos'])

plt.show()
```

```python
z = np.cos(x)

plt.subplot(2, 2, 1)    # 2 rows, 2 cols, index = 1
plt.plot(x, y, color='red')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Titlu 1')
plt.legend(['Sin', 'Cos'])

plt.subplot(2, 2, 2)    # al 2lea plot
plt.plot(x, z, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Titlu 2')
plt.legend(['Sin', 'Cos'])
 
plt.subplot(2, 2, 3)
plt.plot(x, y, color='red')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Titlu 1')
plt.legend(['Sin', 'Cos'])

plt.subplot(2, 2, 4)
plt.plot(x, z, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Titlu 2')
plt.legend(['Sin', 'Cos'])

plt.show()
```

```python
# Ex 1
# pip install scikit-image

from zipfile import ZipFile
zf = ZipFile('images.zip', 'r')
zf.extractall()
zf.close()

import os
files = os.listdir('images')
print(files)

# a)
images = [] # lista cu imagini

for i in range(len(files)):
    img_name = files[i]
    image = np.load('images/' + img_name)
    images.append(image)
    
# salvez imaginile intr-un np.array
np_images = np.array(images)
print(np_images.shape)


for i in range(len(files)):
    plt.subplot(3, 3, i + 1) # pozitia i + 1
    plt.imshow(np_images[i], cmap = 'gray')
```

```python
# b)
pixelsSum = np.sum(np_images)
print(pixelsSum)
```

```python
# c)
sumPerImage = np.sum(np_images, axis = (1, 2))
print(sumPerImage)   

# sau 
for i in range(9):
    print(np.sum(np_images[i, :, :]))
```

```python
# d) 
idxMax = np.argmax(sumPerImage)
print(idxMax)
```

```python
# e) imaginea medie + afisare
# pip install scikit-image

mean_img = np.mean(np_images, axis = 0)
plt.imshow(mean_img.astype(np.uint8), cmap = 'gray')
plt.show()
```

```python
# f) deviatia standard a imaginilor
std = np.std(np_images)
print(std)
```

```python
# g) normalizare (se scade imaginea medie i se mparte rezultatul la deviaia standard)
normalised = (np_images - mean_img) / std
# print(normalised)

print(np.mean(images), np.std(images))
print(np.mean(normalised), np.std(normalised))
```
